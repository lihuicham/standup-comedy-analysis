```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import *

def weighted_predictions(full_model, filtered_model, X, weight = 0.7):
    full_probs = full_model.predict_proba(X)
    filtered_probs = filtered_model.predict_proba(X)

    adj_probs = (full_probs * weight) + (filtered_probs * (1 - weight))
    y_hat = adj_probs.argmax(axis = 1)
    return y_hat


weighted_pred = weighted_predictions(
    full_grad_clf, filtered_grad_clf, X_train_sm[full_features]
)
accuracy_score(y_train_sm, weighted_pred)
```

```python
from sklearn.model_selection import train_test_split

full_X = valid_transcripts_sent_df["Processed Transcript"]
full_y = valid_transcripts_sent_df["Funniness"]

full_X = pd.DataFrame(
    full_vectorizer.transform(full_X).toarray(), 
    columns = full_vectorizer.get_feature_names_out()
)[full_features]

X_train, X_test, y_train, y_test = train_test_split(full_X, full_y, test_size = 0.3, random_state = 242)
```

```python
def kfold_cv(full_model, filtered_model, X, y, params, k = 5):
    partition_idx = np.linspace(0, len(X), k + 1).astype(int)
    
    cv_results = []
    for weight in params:
        cv_scores = []
        for idx in range(k):
            start = partition_idx[idx]
            end = partition_idx[idx + 1]

            X_holdout = X[start: end]
            X_training = pd.concat([X[0: start], X[end :]])

            y_holdout = y[start: end]
            y_training = pd.concat([y[0: start], y[end :]])

            full_prob = full_model.predict_proba(X_training)
            filtered_prob = filtered_model.predict_proba(X_training)

            weighted_prob = (full_prob * weight) + (filtered_prob * (1 - weight))
            weighted_pred = np.argmax(weighted_prob, axis = 1)
            
            roc = roc_auc_score(
                pd.get_dummies(y_training),
                pd.get_dummies(weighted_pred),
                multi_class = "ovo"
            )
            cv_scores.append((roc, weight))
        cv_results.append(cv_scores)

    return cv_results

params = np.linspace(0, 1, 5)
cv_results = kfold_cv(full_grad_clf, filtered_grad_clf, X_train, y_train, params)

avg_rocs = []
for weight_fold in cv_results:
    avg_roc = np.mean([fold[0] for fold in weight_fold])
    avg_rocs.append(avg_roc)

best_param = params[np.argmax(avg_rocs)]
best_param
```

```python
weighted_pred = weighted_predictions(full_grad_clf, filtered_grad_clf, X_train, weight = best_param)
full_pred = full_grad_clf.predict(X_train)
filtered_pred = filtered_grad_clf.predict(X_train)

weighted_roc = roc_auc_score(
    pd.get_dummies(y_train),
    pd.get_dummies(weighted_pred),
    multi_class = "ovo"
)
full_roc = roc_auc_score(
    pd.get_dummies(y_train),
    pd.get_dummies(full_pred),
    multi_class = "ovo"
)
filtered_roc = roc_auc_score(
    pd.get_dummies(y_train),
    pd.get_dummies(filtered_pred),
    multi_class = "ovo"
)

pd.DataFrame(dict(
    ROC = [weighted_roc, full_roc, filtered_roc],
), index = ["Weighted", "Full", "Filtered"])
```
