<h1>Import Libraries</h1>

```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
```

<h1>Scrape Comedy Transcript</h1>

```python
def scrapeURL(url):

    # Downloading page's HTML
    res = requests.get(url)

    # Passing HTML through BeautifulSoup constructor
    soup = BeautifulSoup(res.content, "html.parser")

    # Finding div containing text transcript
    transcript = soup.find("div", class_="elementor-widget-theme-post-content")

    return transcript.div.get_text()
```

```python
# Test function with input of URL

scrapeURL("https://scrapsfromtheloft.com/comedy/chris-rock-selective-outrage-transcript/")
```

<h1>Navigate to Transcript Site</h1>

```python
def scrapeByArtist(comedian):
    
    # Get HTMl of page
    res = requests.get("https://scrapsfromtheloft.com/stand-up-comedy-scripts/")
    
    # Passing HTML through BeautifulSoup constructor
    soup = BeautifulSoup(res.content, "html.parser")
    
    # Parse HTML response content using LXML parser
    dom = etree.HTML(str(soup))
    
    # Find div containers for all URLs
    divs = dom.xpath('//*[@id="post-19174"]/div/div/div/div/section[6]/div/div/div/div/div/div/div/div/article/div/h3/a')
    
    # Filter div containers for URLs that matches the input comedian
    def match_by_comedian(div):    
        
        # Extract comedian name from div of URL
        comedian_name = div.text.split(':')[0].strip()
        
        return comedian_name == comedian
    
    filtered_by_comedian = filter(match_by_comedian, divs)
    
    # Construct dataframe
    df = pd.DataFrame()

    # Populate dataframe with program of comedian, and scraped transcript
    for filtered_div in filtered_by_comedian:
        url = filtered_div.get('href')
        df = df.append({'Program' : filtered_div.text.strip(), 'Transcript' : scrapeURL(url)},
                 ignore_index = True)

    return df
```

```python
# Test function with input of comedian

scrapeByArtist("Chris Rock")
```
