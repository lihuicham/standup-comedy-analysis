```python
## importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get('https://scrapsfromtheloft.com/stand-up-comedy-scripts/')
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
comedian_col = []
title_col = []
transcript_col = []

transcripts_h3 = soup.find_all('h3', class_='elementor-post__title')[4:]

for t in transcripts_h3:
    href = t.find('a').get('href')
    t_req = requests.get(href)
    t_soup = BeautifulSoup(t_req.content, 'html.parser')
    
    comedian_and_title = t_soup.find('h1', class_='elementor-heading-title elementor-size-default').get_text()
    comedian_title_list = comedian_and_title.split(': ')
    if len(comedian_title_list) == 2:
        comedian = comedian_and_title.split(': ')[0]
        comedian_col.append(comedian)
        title = comedian_and_title.split(': ')[1]
        title_col.append(title)
    elif len(comedian_title_list) == 1:
        title_col.append(comedian_and_title)
        comedian_col.append('NA')
    elif len(comedian_title_list) == 3 and comedian_title_list[0] == 'Dave Chappelle' and comedian_title_list[2] == 'The Punchline – Transcript':
        comedian_col.append(comedian_title_list[0])
        title_col.append(': '.join(comedian_title_list[1:]))
    print(comedian_and_title)
    
    paras = t_soup.find_all('p', attrs={"style": "text-align: justify;"})
    para_list = [para.get_text() for para in paras]
    para_text = '\n'.join(para_list)
    transcript_col.append(para_text)
```

```python
df_dic = {'Comedian': comedian_col, 'Title': title_col, 'Transcript': transcript_col}
df = pd.DataFrame(df_dic)
df.to_csv('transcripts.csv')
df
```
